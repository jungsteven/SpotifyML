```python
import pandas as pd
from sklearn import tree
#from sklearn import DecisionTreeRegressor
import os
import numpy
```

```python
df = pd.read_csv('data/master_dataset_float.csv', encoding="ISO-8859-1")
df.head(10)
```

```python
X = df[["energy", "instrumentalness", "liveness", "loudness", "speechiness", "tempo"]]
y = df["danceability"].values.reshape(-1, 1)
#y = df["danceability"]
print(X.shape, y.shape)    
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head(10)
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
